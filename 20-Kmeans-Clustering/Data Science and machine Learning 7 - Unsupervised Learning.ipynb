```python
#Total ouliers found
np.sum(dbscan.labels_ == -1)
```

```python
#percent of points classified as outliers
100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_)
```

```python
np.unique(dbscan.labels_)
```

```python
outlier_percent = []
number_of_outliers = []


for eps in np.linspace(0.001,10,100):
    
    dbscan = DBSCAN(eps=eps)
    dbscan.fit(two_blobs_outliers)
    
    #Total outlier found
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    #Percentage outliers found
    outlier_percent.append(100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_))
```

```python
sns.lineplot(x=np.linspace(0.001,10,100),y = number_of_outliers)
```

```python
sns.lineplot(x=np.linspace(0.001,10,100),y = number_of_outliers)
plt.xlim(0,1.5)
plt.ylim(0,10)
plt.hlines(y=3,xmin=0,xmax=2,color='red')

## - - > 0.75 seems like a good inflection point. 
# again this is based on having an intuition of what the no of outliers should be
```

```python
#percent of points considered as outliers
sns.lineplot(x=np.linspace(0.001,10,100),y = outlier_percent)
plt.xlim(0,0.6)
plt.hlines(y=3,xmin=0,xmax=2,color='red')
```

```python
outlier_percent = []
number_of_outliers = []


for n in np.arange(1,100):
    
    dbscan = DBSCAN(min_samples=n)
    dbscan.fit(two_blobs_outliers)
    
    #Total outlier found
    number_of_outliers.append(np.sum(dbscan.labels_ == -1))
    
    #Percentage outliers found
    outlier_percent.append(100 * np.sum(dbscan.labels_ == -1) / len(dbscan.labels_))
```

```python
sns.lineplot(x=np.arange(1,100),y = outlier_percent)
plt.xlabel('Min number of points for a core point')
plt.ylabel('percentage of points classififed as an outlier')
#plt.xlim(0,1.5)
#plt.ylim(0,10)
#plt.hlines(y=3,xmin=0,xmax=2,color='red')
```

```python
#Good starting point for min samples is 2 times dimension of data

num_dim = two_blobs_outliers.shape[1]

dbscan = DBSCAN(min_samples=num_dim, eps=0.7) 
display_categories(dbscan,two_blobs_outliers)
```

```python
df = pd.read_csv('../DATA/cancer_tumor_data_features.csv')
```

```python
df.head()
```

```python
sns.heatmap(df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaledX = scaler.fit_transform(df)
```

```python
scaledX.mean(axis=0)
```

```python
scaledX - scaledX.mean(axis=0)
```

```python
#grab covariance matrix
```

```python
covariance_matrix = np.cov(scaledX,rowvar=False)
```

```python
eigen_values,eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors.shape
```

```python
#so far from 
# PCA N feature space --? N PC Space
# How many PC to choose? Lets fo with N = 2
# so from PCA 30 feature space --? 30 PC Space --> 2
```

```python
num_components = 2
```

```python
#ans = [0,1,2]
np.argsort([2,1,3])
```

```python
#np.argsort(eigen_values)#[::-1]#[:num_components]
```

```python
#np.argsort(eigen_values)[::-1][:num_components]
```

```python
sorted_key = np.argsort(eigen_values)[::-1][:num_components]
```

```python
eigen_values,eigen_vectors=eigen_values[sorted_key],eigen_vectors[:,sorted_key]
```

```python
#original data -> project -> eigen vector
principal_components = np.dot(scaledX,eigen_vectors)
```

```python
principal_components#[:,1]
```

```python
sns.scatterplot(principal_components[:,0],principal_components[:,1])
```

```python
df.head()
```

```python
scaler = StandardScaler()
```

```python
Xscale = scaler.fit_transform(df)
```

```python
len(Xscale)
```

```python
Xscale
```

```python
from sklearn.decomposition import PCA

#use pca model to reduce feature space/dimensionality or # NOTE: Scaled data
```

```python
help(PCA)
```

```python
pca_model = PCA(n_components=2)
```

```python
pca_model.fit_transform(Xscale) 
#fit calcs the new avg data midpoint,
#transformation does the repositioning/transforming of the data numbers/features 
#...based on this new avg point
```

```python
pc_results = pca_model.fit_transform(Xscale) 
```

```python
plt.figure(dpi=150)
plt.scatter(pc_results[:,0],pc_results[:,1],c = cancer_dictionary['target'])
```

```python
#taking a look at the model

pca_model.components_
```

```python
df_comp = pd.DataFrame(pca_model.components_,index=['PC1','PC2'],columns=df.columns)
```

```python
df_comp #direction of max variance of the data
```

```python
plt.figure(dpi=200,figsize=(20,3))
sns.heatmap(df_comp,annot=True)
```

```python
## very useful - how the diff components planes explain variance in data

pca_model.explained_variance_ratio_
```

```python
np.sum(pca_model.explained_variance_ratio_) 
#63% of data explained by these 2 planes
#with 63% of variance, still able to separate the classes quite well
```

```python
pca_30 = PCA(n_components=30)
```

```python
pca_30.fit(Xscale) #no need transform bcus here only need model params not the data
```

```python
pca_30.explained_variance_ratio_
```

```python
np.sum(pca_30.explained_variance_ratio_)
#100% of all variance accounted here as all data is present
```

```python
explained_variance = []

for n in range(1,30):
    pca = PCA(n_components=n)
    pca.fit(Xscale)
    rat = pca.explained_variance_ratio_
    sumrat = np.sum(rat)*100
    explained_variance.append(sumrat)
```

```python
explained_variance
```

```python
plt.figure(dpi=150)
plt.xlabel('No of components')
plt.ylabel('Variance explanation')
sns.lineplot(range(1,30),explained_variance,marker='o')
```
