```python
sorted_ratings = user_items.sort_values('timestamp')
training_data = sorted_ratings.head(sorted_ratings.shape[0]-100000)
testing_data = sorted_ratings.tail(100000)

# create user-by-item matrix for training and testing groups
full_matrix = user_items.groupby(['user_id', 'movie_id'])['rating'].max().unstack()
training_matrix = training_data.groupby(['user_id', 'movie_id'])['rating'].max().unstack()
testing_matrix = testing_data.groupby(['user_id', 'movie_id'])['rating'].max().unstack()

training_matrix.shape, testing_matrix.shape, full_matrix.shape

#def create_sparse(user_items):
#    return sparse.csr_matrix((user_items.rating, (user_items.user_id, user_items.movie_id)))
#    
#sprs_ratings = create_sparse(user_items)
#sprs_training = create_sparse(training_data)
#sprs_testing = create_sparse(testing_data)
print(testing_data.shape[0] - sprs_testing.nonzero()[0].size)
print(training_data.shape[0] - sprs_training.nonzero()[0].size)
print(training_data.rating.isnull().sum())
print(testing_data.rating.isnull().sum())
print(testing_data.shape)
print(training_data.shape)
print(full_contents.shape[0] - sprs_ratings.nonzero()[0].size)
print(sprs_ratings.nonzero()[0].size)
print(full_contents.shape[0])

not np.isnan(training_np[0,0])

indices_of_nan = np.argwhere(np.isnan(training_np))

latent_features = 4
n_users = ratings_mat.shape[0]
n_movies = ratings_mat.shape[1]
n_already_rated = ratings_mat.nonzero()[0].size
user_mat = np.random.rand(n_users*latent_features).reshape([n_users, latent_features])
movie_mat = np.random.rand(n_movies*latent_features).reshape([latent_features, n_movies])


def FunkSVD_w_regularization(ratings_mat, latent_features=4, learning_rate=0.005, regularization=0.02, iters=100):
    '''
    What does this function do?
    
    INPUT:
    
    OUTPUT:
    
    '''
    n_users = ratings_mat.shape[0]
    n_movies = ratings_mat.shape[1]
    n_already_rated = ratings_mat.nonzero()[0].size
    user_mat = np.random.rand(n_users, latent_features)
    movie_mat = np.random.rand(latent_features, n_movies)

    iter_count = 0
    sse_accum = 0
    print("Optimizaiton Statistics")
    print("Iterations | Mean Squared Error ")
    for i in range(iters):
        old_sse = sse_accum
        sse_accum = 0
        
        # For each user-movie pair
        for i in range(n_users):
            for j in range(n_movies):
                
                # if the rating exists
                if ratings_mat[i, j] > 0:
                    
                    # compute the error as the actual minus the dot product of the user and movie latent features
                    diff = ratings_mat[i, j] - np.dot(user_mat[i, :], movie_mat[:, j])
                    
                    # Keep track of the sum of squared errors for the matrix
                    sse_accum += diff**2
                    
                    # update the values in each matrix in the direction of the gradient regularize by how large the value was to begin with
                    for k in range(latent_features):
                        user_mat[i, k] += learning_rate * (2*diff*movie_mat[k, j] - regularization*user_mat[i, k])
                        movie_mat[k, j] += learning_rate * (2*diff*user_mat[i, k] - regularization*movie_mat[k, j])

        # print results
        print("%d \t\t %f" % (iter_count, sse_accum/n_already_rated))
        
        # update iterations
        iter_count += 1
        
    return user_mat, movie_mat
```
