## A Linear Regression model that can predict prices on an advertising data set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sklearn
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
df['total_spend'] = df['TV'] +df['radio']+df['newspaper']
```

```python
df.head()
```

```python
sns.scatterplot(data=df,x='total_spend',y='sales')
```

```python
sns.regplot(data=df,x='total_spend',y='sales')
```

```python
x = df['total_spend']
y = df['sales']
```

```python
np.polyfit(x,y,deg=1) #deg = 1, for considering only x, nothing like x**2
```

```python
potential_spend = np.linspace(0,500,100)
```

```python
predicted_sales = 0.04868788*potential_spend + 4.24302822
```

```python
sns.scatterplot(data=df,x='total_spend',y='sales')
plt.plot(potential_spend,predicted_sales,color = 'red')
```

```python
spend = 200

predicted_sales = 0.04868788*spend + 4.24302822
```

```python
predicted_sales
```

```python
np.polyfit(x,y,3)
```

```python
poly = np.polyfit(x,y,3)
poly[3]
```

```python
# y = B1x + B0
# y = B3x**3 + B2x**2 ++ B1x + B0
```

```python
pot_spend = np.linspace(0,500,100)
```

```python
pred_sales = poly[0]*pot_spend**3 + poly[1]*pot_spend**2+ poly[2]*pot_spend**1+ poly[3]
```

```python
pred_sales
```

```python
sns.scatterplot(data=df,x='total_spend',y='sales')
plt.plot(pot_spend,pred_sales,color = 'red')
```

```python
df = df.drop('total_spend',axis=1)
```

```python
df.head()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].set_title("Newspaper Spend");
axes[2].set_ylabel("Sales")
plt.tight_layout();
```

```python
sns.pairplot(df,corner=True)
```

```python
X = df.drop('sales',axis=1)
```

```python
X.head()
```

```python
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
len(df)
```

```python
len(X_train)
```

```python
#y_train
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#help(LinearRegression)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
test_predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
df['sales'].mean()
```

```python
sns.histplot(data=df,x='sales')
```

```python
mean_absolute_error(y_test,test_predictions)
```

```python
mean_squared_error(y_test,test_predictions)
```

```python
np.sqrt(mean_squared_error(y_test,test_predictions))
```

```python
# evaluate residuals
```

```python
test_residuals = y_test - test_predictions
```

```python
#test_residuals
```

```python
sns.scatterplot(x=y_test,y=test_residuals)
plt.axhline(y=0,color='red',ls='--')
```

```python
sns.displot(test_residuals,bins=25,kde=True)
```

```python
final_model = LinearRegression()
```

```python
final_model.fit(X,y) 
```

```python
final_model.coef_ #[2]
```

```python
X.head()
```

```python
y_hat = final_model.predict(X)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))

axes[0].plot(df['TV'],df['sales'],'o')
axes[0].plot(df['TV'],y_hat,'o',color='red')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['radio'],df['sales'],'o')
axes[1].plot(df['radio'],y_hat,'o',color='red')
axes[1].set_title("Radio Spend")
axes[1].set_ylabel("Sales")

axes[2].plot(df['newspaper'],df['sales'],'o')
axes[2].plot(df['radio'],y_hat,'o',color='red')
axes[2].set_title("Newspaper Spend");
axes[2].set_ylabel("Sales")
plt.tight_layout();
```

```python
from joblib import dump,load
```
