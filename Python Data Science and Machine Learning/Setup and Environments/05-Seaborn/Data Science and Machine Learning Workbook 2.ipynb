```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df.head()
```

```python
plt.figure(figsize = (12,4),dpi=200)

sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize = (12,8),dpi=200)

sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',
                palette = 'Dark2')
```

```python
plt.figure(figsize = (12,8),dpi=200)

sns.scatterplot(x='salary',y='sales',data=df,size='work experience')
```

```python
plt.figure(figsize = (12,8),dpi=200)

sns.scatterplot(x='salary',y='sales',data=df,s=200,alpha =0.2)
```

```python
plt.figure(figsize = (12,8),dpi=200)

sns.scatterplot(x='salary',y='sales',data=df,s=200,style='level of education',
               hue='level of education',alpha=0.9)

#plt.savefig("my_plot.jpg")
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
plt.figure(dpi=200)
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
#DO NOT USE DISTPLOT -deprecated

sns.set(style='darkgrid') #whitegrid, white,dark,tick - high level cmd, set for whole notebook
sns.displot(data=df,x='salary',bins=10,color='red',edgecolor='blue'
           ,lw=1.5,ls='--')
```

```python
sns.set(style='darkgrid') #whitegrid, white,dark,tick - high level cmd, set for whole notebook
sns.histplot(data=df,x='salary',bins=10,color='red')
```

```python
sns.set_style('white')
sns.displot(data=df,x='salary',bins=10,kde=True,rug=True)
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages=pd.DataFrame(sample_ages,columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages,x='age')
```

```python
sns.displot(data=sample_ages,x='age',rug=True,bins=20,kde=True)
```

```python
sns.kdeplot(data=sample_ages,x='age')
```

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100]) #trim the ranges of the plot
```

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100],bw_adjust=0.1) 
#bw - to pick up more of the probability variance between points
```
