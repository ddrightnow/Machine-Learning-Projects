```python
X = pd.get_dummies(df)
```

```python
X.head() #dont bothe about correlating columns for K Means clustering
```

```python
#scale our data for distance related features
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2) #arbitratry, ppl who want to market to, ppl we dont want to 
```

```python
#fit, then predict; no separationg of fit and predict since this is: 
#UNSUPERVIUSDE LEARNING
cluster_labels = model.fit_predict(scaled_X)
```

```python
cluster_labels
```

```python
X['Cluster'] = cluster_labels
```

```python
X.head()
```

```python
X.corr()
```

```python
plt.figure(dpi=150,figsize=(12,6))
X.corr()['Cluster'].iloc[:-1].sort_values().plot(kind='bar')
```

```python
#Methodologies for choosing a K value
```

```python
ssd = []

for k in range(2,10):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)
    
    ssd.append(model.inertia_) #sum of squared distance/ssd -->cluster center
```

```python
ssd
```

```python
plt.plot(range(2,10),ssd,marker='s',ls='--')
```

```python
pd.Series(ssd)
```

```python
pd.Series(ssd).diff() #calc the drop from one cell to the next
```
