## The Data

A tunnel boring company  uses X-rays in an attempt to know rock density, ideally this will allow them to switch out boring heads on their equipment before having to mine through the rock!

<img src='boring.jpg' height=400>

They have given us some lab test results of signal strength returned in nHz to their sensors for various rock density types tested. The dataset has an almost sine wave like relationship, where signal strength oscillates based off the density, the researchers are unsure why this is.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/rock_density_xray.csv")
```

```python
df.head()
```

```python
df.columns=['Signal',"Density"]
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(x='Signal',y='Density',data=df)
```

---
---

## Splitting the Data


```python
X = df['Signal'].values.reshape(-1,1)  
y = df['Density']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

-----

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
lr_preds = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test,lr_preds))
```

```python
signal_range = np.arange(0,100)
```

```python
lr_output = lr_model.predict(signal_range.reshape(-1,1))
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(x='Signal',y='Density',data=df,color='black')
plt.plot(signal_range,lr_output)
```

# Polynomial Regression

## Attempting with a Polynomial Regression Model

## Function to Help Run Models

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
def run_model(model,X_train,y_train,X_test,y_test):
    
    # Fit Model
    model.fit(X_train,y_train)
    
    # Get Metrics
    
    preds = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test,preds))
    print(f'RMSE : {rmse}')
    
    # Plot results
    signal_range = np.arange(0,100)
    output = model.predict(signal_range.reshape(-1,1))
    
    
    plt.figure(figsize=(12,6),dpi=150)
    sns.scatterplot(x='Signal',y='Density',data=df,color='black')
    plt.plot(signal_range,output)
```

```python
run_model(model,X_train,y_train,X_test,y_test)
```

## Pipeline for Poly Orders

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pipe = make_pipeline(PolynomialFeatures(2),LinearRegression())
```

```python
run_model(pipe,X_train,y_train,X_test,y_test)
```

## Comparing Various Polynomial Orders

```python
pipe = make_pipeline(PolynomialFeatures(10),LinearRegression())
run_model(pipe,X_train,y_train,X_test,y_test)
```

## KNN Regression

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
preds = {}
k_values = [1,5,10]
for n in k_values:
    
    
    model = KNeighborsRegressor(n_neighbors=n)
    run_model(model,X_train,y_train,X_test,y_test)
```

---
## Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor()

run_model(model,X_train,y_train,X_test,y_test)
```

```python
model.get_n_leaves()
```

----

## Support Vector Regression

```python
from sklearn.svm import SVR
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.01,0.1,1,5,10,100,1000],'gamma':['auto','scale']}
svr = SVR()
```

```python
grid = GridSearchCV(svr,param_grid)
```

```python
run_model(grid,X_train,y_train,X_test,y_test)
```

```python
grid.best_estimator_
```

## Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# help(RandomForestRegressor)
```

```python
trees = [10,50,100]
for n in trees:
    
    model = RandomForestRegressor(n_estimators=n)
    
    run_model(model,X_train,y_train,X_test,y_test)
```
